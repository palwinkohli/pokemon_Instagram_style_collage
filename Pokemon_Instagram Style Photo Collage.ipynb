```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Read images
bottom_left = cv2.imread('Downloads/ImageCollage dataset/bottom_left.jpg')
plt.imshow(bottom_left)
```

```python
bottom_right = cv2.imread('Downloads/ImageCollage dataset/bottom_right.jpg')
plt.imshow(bottom_right)
```

```python
center = cv2.imread('Downloads/ImageCollage dataset/center.jpeg')
plt.imshow(center)
```

```python
top_left = cv2.imread('Downloads/ImageCollage dataset/top_left.jpg')
plt.imshow(top_left)
```

```python
top_right = cv2.imread('Downloads/ImageCollage dataset/top_right.jpg')
plt.imshow(top_right)
```

```python
#Convert to RGB
rgb_image_bottom_left = cv2.cvtColor(bottom_left, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image_bottom_left)
```

```python
rgb_image_bottom_right = cv2.cvtColor(bottom_right, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image_bottom_right)
```

```python
rgb_image_center = cv2.cvtColor(center, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image_center)
```

```python
rgb_image_top_left = cv2.cvtColor(top_left, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image_top_left)
```

```python
rgb_image_top_right = cv2.cvtColor(top_right, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image_top_right)
```

```python
#Resize and add padding
rgb_image_bottom_left = cv2.resize(rgb_image_bottom_left,(200,200))
rgb_image_bottom_left = cv2.copyMakeBorder(rgb_image_bottom_left, 5, 10, 10, 5, cv2.BORDER_CONSTANT)

rgb_image_bottom_right = cv2.resize(rgb_image_bottom_right,(200,200))
rgb_image_bottom_right = cv2.copyMakeBorder(rgb_image_bottom_right, 5, 10, 5, 10, cv2.BORDER_CONSTANT)

rgb_image_center = cv2.resize(rgb_image_center,(100,100))
rgb_image_center = cv2.copyMakeBorder(rgb_image_center, 10, 10, 10, 10, cv2.BORDER_CONSTANT)

rgb_image_top_left = cv2.resize(rgb_image_top_left,(200,200))
rgb_image_top_left = cv2.copyMakeBorder(rgb_image_top_left, 10, 5, 10, 5, cv2.BORDER_CONSTANT)

rgb_image_top_right = cv2.resize(rgb_image_top_right,(200,200))
rgb_image_top_right = cv2.copyMakeBorder(rgb_image_top_right, 10, 5, 5, 10, cv2.BORDER_CONSTANT)
```

```python
#Concatenate images
top_section = np.concatenate((rgb_image_top_left,rgb_image_top_right), axis = 1)
plt.imshow(top_section)
```

```python
bottom_section = np.concatenate((rgb_image_bottom_left,rgb_image_bottom_right), axis = 1)
plt.imshow(bottom_section)
```

```python
top_bottom = np.concatenate((top_section,bottom_section), axis = 0)
plt.imshow(top_bottom)
plt.axis("On")
```

```python
#Impose the 5th image at the center
top_bottom[155:275,155:275] = rgb_image_center
plt.imshow(top_bottom)
```

```python
#Convert back to BGR and then save as jpg
top_bottom_bgr = cv2.cvtColor(top_bottom,cv2.COLOR_RGB2BGR)
cv2.imwrite('pokemon_instagram_collage.jpg',top_bottom_bgr)
```

```python
#Convert to array, reshape and save as .csv
collage_array = np.array(top_bottom)
new_img = top_bottom.reshape((top_bottom.shape[0]*top_bottom.shape[1]), top_bottom.shape[2])
np.savetxt("Pokemon_instagram_collage.csv",new_img, header = 'r,g,b',fmt='% 4d',delimiter=',')
```
